```python
heart = pd.read_csv("Heart_Disease_Prediction (1).csv")
heart.head()
```

```python
#Create Age Groups
#Age is continuous. Grouping it into ranges may help some models and also makes it easier to interpret.
heart["Age_Group"] = pd.cut(
    heart["age"],
    bins=[0, 30, 40, 50, 60, 100],
    labels=[
        "18-30",
        "31-40",
        "41-50",
        "51-60",
        "61+"
    ]
)
heart["Age_Group"].value_counts()
```

```python
heart["Chol_Category"] = pd.cut(
    heart["Cholesterol"],
    bins=[0, 199, 239, 700],
    labels=[
        "Normal",
        "Borderline",
        "High"
    ]
)
heart["Chol_Category"].value_counts()
```

```python
#Blood Pressure Category
#Using resting blood pressure:
heart["BP_Category"] = pd.cut(
    heart["BP"],
    bins=[0, 120, 129, 139, 300],
    labels=[
        "Normal",
        "Elevated",
        "Stage1",
        "Stage2"
    ]
)
heart["BP_Category"].value_counts()
```

```python

#Interaction Features
'''These combine two existing variables.'''

#Age × Cholesterol'''
heart["Age_Chol"] = heart["Age"] * heart["Cholesterol"]
#'''Age × Blood Pressure'''
heart["Age_BP"] = heart["Age"] * heart["BP"]
#Cholesterol × Blood Pressure
heart["Chol_BP"] = heart["Cholesterol"] * heart["BP"]
```

```python
#Heart Rate Ratio
heart["HeartRate_Age"] = heart["Max HR"] / heart["Age"]
```

```python
#Verify New Features
heart.head()
```

```python
#Check for Missing Values
heart.isnull().sum()
```

```python
#Save the Feature-Engineered Dataset
heart.to_csv("heart_disease_feature_engineered.csv", index=False)
```

```python
heart = pd.read_csv("heart_disease_feature_engineered.csv")

heart.head()
```

```python
#Check Data Types
heart.dtypes
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

heart["Age_Group"] = encoder.fit_transform(heart["Age_Group"])

heart["Chol_Category"] = encoder.fit_transform(heart["Chol_Category"])

heart["BP_Category"] = encoder.fit_transform(heart["BP_Category"])
```

```python
heart.head()
```

```python
#Define X and y
# Remove:

# target
X = heart.drop(columns=["Heart Disease"])

y = heart["Heart Disease"]
print(X.shape)
print(y.shape)
```

```python
#Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
print(X_train.shape)
print(X_test.shape)
```

```python
#Feature Scaling
'''This is important for:

SVM
KNN
Logistic Regression'''
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
#Check Class Distribution
print(y_train.value_counts())
```

```python
#Save the Preprocessed Data
import joblib

joblib.dump(scaler, "scaler.pkl")
```

```python
#Import Libraries
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

```python
#Install Libraries
!pip install xgboost lightgbm
```

```python
#Import Them
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
```

```python
#Create All Models
models = {

    "Logistic Regression":
    LogisticRegression(random_state=42),

    "Decision Tree":
    DecisionTreeClassifier(random_state=42),

    "Random Forest":
    RandomForestClassifier(random_state=42),

    "SVM":
    SVC(probability=True, random_state=42),

    "KNN":
    KNeighborsClassifier(),

    "XGBoost":
    XGBClassifier(
        random_state=42,
        eval_metric="logloss"
    ),

    "LightGBM":
    LGBMClassifier(random_state=42)

}
```

```python
results = []

for name, model in models.items():

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    y_prob = model.predict_proba(X_test)[:, 1]

    results.append({

        "Model": name,

        "Accuracy":
        accuracy_score(y_test, y_pred),

        "Precision":
        precision_score(y_test, y_pred, pos_label='Presence'),

        "Recall":
        recall_score(y_test, y_pred, pos_label='Presence'),

        "F1 Score":
        f1_score(y_test, y_pred, pos_label='Presence'),

        "ROC AUC":
        roc_auc_score(y_test, y_prob)

    })
```

```python
#Display Results
results_df = pd.DataFrame(results)

results_df
```

```python
#Sort by Accuracy
results_df.sort_values(
    by="Accuracy",
    ascending=False
)
```

```python
#Import GridSearchCV
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMClassifier
```

```python
#Create the Model
lgbm = LGBMClassifier(random_state=42)
```

```python
#Define the Parameter Grid
param_grid = {
    "n_estimators": [100, 200],
    "learning_rate": [0.01, 0.05, 0.1],
    "max_depth": [3, 5, 7],
    "num_leaves": [15, 31, 50]
}
```

```python
#Perform Grid Search
grid_search = GridSearchCV(
    estimator=lgbm,
    param_grid=param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
#Best Parameters
print("Best Parameters:")
print(grid_search.best_params_)
```

```python
#Best Cross-Validation Accuracy
print("Best CV Accuracy:", grid_search.best_score_)
```

```python
#Test the Tuned Model
best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)
```

```python
#Calculate the New Accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy after tuning: {accuracy*100:.2f}%")
```
